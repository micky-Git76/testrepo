```python
import docx2txt
my_text=docx2txt.process("discoursKM3.docx")
stopwords = set(STOPWORDS)
type(my_text)
```

```python
# instantiate a word cloud object
KM_wc = WordCloud(
    background_color='white',
    max_words=15000,
    stopwords=stopwords
)

# generate the word cloud
stopwords.add('will') # add the words said to stopwords
stopwords.add('ve')
stopwords.add('time')
stopwords.add('res')
stopwords.add('s')
stopwords.add('re')
stopwords.add('week')
stopwords.add('going')
stopwords.add('whilst')
stopwords.add('look')
stopwords.add('two')
stopwords.add('start')
stopwords.add('day')
stopwords.add('last')
stopwords.add('much')
stopwords.add('couple')
KM_wc.generate(my_text)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(14, 18))

# display the cloud
plt.imshow(KM_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
